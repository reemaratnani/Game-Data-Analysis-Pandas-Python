```python
import pandas as pd
import numpy as np
import os
# os.getcwd()

```

```python
df_original = pd.read_csv("purchase_data.csv")
# df_original.head()
```

```python
players_count = len(df_original['SN'].unique())
# print(players_count)
Total_players_df = pd.DataFrame({
    "Total Players": [players_count]
})
Total_players_df


```

```python
count_items = len(df_original['Item ID'].unique())
# print(count_items)
Average_price = df_original['Price'].mean()
Average_Price = round(Average_price, 2)
Number_purchase = len(df_original['Purchase ID'])
# print(Number_purchase)
Total_Revenue = df_original['Price'].sum()
# print(Total_Revenue )
```

```python
Purchasing_summary = pd.DataFrame({
                                  "Number of Unique Items": [count_items],
                                  "Average Price": [Average_price],
                                  "Number of Purchases": [Number_purchase],
                                  "Total Revenue": [Total_Revenue]
                                  })
Purchasing_summary
```

```python
gender_category = df_original[['SN','Gender']]
# gender_category.head()
```

```python
gender_list = gender_category.drop_duplicates(['SN'])
# gender_list.head()
gender_demo = gender_list['Gender'].value_counts()
gender_df = pd.DataFrame(gender_demo)
gender_df = gender_df.rename(columns={"Gender":"Total Count"})
gender_df
```

```python
percentage_player = gender_df['Total Count']*100/gender_df['Total Count'].sum()
gender_df['Percentage of Players'] = percentage_player.map("{:.2f}".format)
gender_df
```

```python
purchase_bygen = df_original[['Purchase ID','Gender', 'Price']]
# purchase_bygen.head()
```

```python
purchase_group = purchase_bygen.groupby("Gender")
summary_df1 = pd.DataFrame({
    "Purchase Count": purchase_group['Purchase ID'].count(),
    "Average Purchase Price": purchase_group['Price'].mean(),
    "Total Purchase Value": purchase_group['Price'].sum(),
    "Average Total Purchase per person": purchase_group['Price'].sum()/gender_df['Total Count']
})

summary_df1['Average Purchase Price'] = summary_df1['Average Purchase Price'].map("${:.2f}".format)
summary_df1
```

Age Demographics


```python
Age_group = df_original['Age']
bins=[0, 9, 14, 19, 24, 29, 34, 39, Age_group.max()]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]
df_original['Age group'] = pd.cut(df_original['Age'], bins, labels=group_labels)
df_original.head()
```

```python
Age_analysis = df_original.drop_duplicates(['SN'])
Total_count = Age_analysis['Age group'].value_counts().sort_index()
Percent_player = Total_count*100/Total_count.sum()
Percent_player = Percent_player.map("{:.2f}".format)
Age_groupdf = pd.DataFrame({"Total Count": Total_count, "Percentage of Players": Percent_player})
Age_groupdf
```

```python
Purchase_df = df_original.groupby('Age group')
Purchase_count = Purchase_df['Purchase ID'].count()
# Purchase_count
Average_price = Purchase_df['Price'].mean()
Total_price = Purchase_df['Price'].sum()
Average_price = Average_price.map("{:.2f}".format)


# Purchase_count = df_original['Age group'].value_counts().sort_index()


Agegroup_df2 = pd.DataFrame({"Purchase Count": Purchase_count,
                             "Average Purchase Price": Average_price,
                             "Total Purchase Value": Total_price,
                             "Average Total purchase per person": Total_price/Age_groupdf['Total Count']
                            })
Agegroup_df2.style.format({"Average Total purchase per person":"{:.2f}"})
```

```python
Spenders_df = df_original.groupby('SN')
# Spenders_df.head()
Spenders_count = Spenders_df['Purchase ID'].count()
# Spenders_count.head(10)

Spenders_avePrice = Spenders_df['Price'].mean()
Spenders_Total = Spenders_df['Price'].sum()
# # Spenders_Total = Spenders_Total.sort_values(ascending=False)
Spenders_summary = pd.DataFrame({
                                 "Purchase Count": Spenders_count,
                                 "Average Purchase Price": Spenders_avePrice,
                                  "Total Purchase Value": Spenders_Total
    
 })
Spenders_summary['Average Purchase Price'] = Spenders_summary['Average Purchase Price'].map("${:.2f}".format)

Spenders_summary= Spenders_summary.sort_values(['Total Purchase Value'],ascending=False)
Spenders_summary.head()
# Spenders_count.sort_values(ascending=False)
```
